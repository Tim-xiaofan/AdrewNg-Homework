```python
import matplotlib.pyplot as plt
import numpy as np;
# tells matplotlib to embed plots within the notebook
%matplotlib inline

sigmas = [1, 0.5, 3]
l = np.array([3, 5]).reshape((2, 1))
ac = 100


for k in range(3):
    fig = plt.figure(k, figsize=(20, 10))
    # surface plot
    x1 = np.linspace(0, 6, ac)
    x2 = np.linspace(0, 10, ac)
    f1 = np.zeros((x1.shape[0], x2.shape[0]))

    # Fill out f1
    sigma = sigmas[k]
    for i, x1_ in enumerate(x1):
        for j, x2_ in enumerate(x2):
            z = np.square(x1_ - l[0]) + np.square(x2_ -l[1])
            z = -z/(2 * sigma )
            f1[i, j] = np.exp(z)
    f1 = f1.T

    x1,x2 = np.meshgrid(x1,x2)
    ax = fig.add_subplot(121, projection='3d')
    ax.plot_surface(x1, x2, f1, cmap='viridis')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('Kernel sigma sqaure={}'.format(sigma))
    ax.plot(l[0], l[1], 1,'ro', ms=20, lw=2)
    
    # contour plot
    # Plot J_vals as 15 contours spaced logarithmically between 0.01 and 100
    ax = plt.subplot(122)
    plt.contour(x1, x2, f1, linewidths=2, cmap='viridis')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.plot(l[0], l[1], 'ro', ms=10, lw=2)
    
# plt.show()
```
